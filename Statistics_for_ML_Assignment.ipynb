# Task-1 Calculating Mean and Median

```python
def calculate_mean(data):
    """Calculate the mean of the data given."""

    if not data:
        return 0
    mean=sum(data)/len(data)
    return mean
def calculate_median(data):
    """Calculates the Median of the data given"""
    if not data:
        return 0
    data_sorted=sorted(data)
    n=len(data_sorted)
    mid=n//2
    if n%2==0 :
        median=(data_sorted[mid-1]+data_sorted[mid])/2
        
    else:
        median=data_sorted[mid]
    return median
```

# Task-2 Calculating Standard Deviation

```python
def calculate_std(data):
    """Calculates the Standard Deviation of the data given"""
    if not data:
        return 0
    
   
    mean=calculate_mean(data)
    variance=sum((x-mean)**2 for x in data)/(len(data))
    std=variance**(0.5)
    return std
```

```python
def display_report(data):
    """Display a report with the mean, median, and standard deviation of the data."""
    print(f"Mean : <{calculate_mean(data)}>")
    print(f"Median : <{calculate_median(data)}>")
    print(f"Standard Deviation : <{calculate_std(data)}>")
```

```python
heights=[170, 165, 180, 172, 168, 160, 175, 171]
display_report(heights)
```
